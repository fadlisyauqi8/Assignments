```python
#Hitung turunan parsial
dx,dy = np.gradient(z)

#Menampilkan ukuran dari turunan
print(" Shape dx : ", dx.shape)
print(" Shape dy : ",dy.shape)
```

```python
#Melakukan perhitungan turunan ordo pertama
print("dx \n", dx)
print("\n dy \n", dy)
```

Setelah berhasil menemukan nilai dari turunan parsian, selanjutnya mencari nilai dari Gradient Magnitude untuk mendeteksi edge pada gambar yang sudah dilakukan filtering.

```python
def mag(x,y):
    mag = []
    for dx,dy in zip(x,y):
        T = np.sqrt((dx*dx)+(dy*dy))
        mag.append(list(T))

    mag = np.array(mag)
    return mag

magnitude = mag(dx,dy)
magnitude
```

```python
#Perhitungan gradient magnitude
magnitude = np.sqrt(dx**2 + dy**2)
```

```python
#Mendeteksi nilai magnitude
image_edge = Image.fromarray(magnitude)  
plt.imshow(image_edge)
```

Dikarenakan pada pengecekan pertama nilai magnitude (Threshold ~ 30), maka harus dicoba untuk menaikkan nilai magnitudenya.

```python
#Mendeteksi batas edge pada magnitude > 40
image_edge = Image.fromarray(magnitude>40)  
plt.imshow(image_edge)
```

Dikarenakan pada batas nilai magnitude diatas 40 membuat hasil deteksi menghilang makan, diharuskan untuk mencoba dibawah nilai 30.

```python
#Mendeteksi batas edge pada magnitude > 20
image_edge = Image.fromarray(magnitude > 20) 
plt.imshow(image_edge)
```

Hasil deteksi gambar pada subject gambar terlihat jelas. selanjutnya kita akan melakukan Transformasi linear untuk rotasi gambar 45 derajat sampai 270 derajat.

```python
rotate_45 = ndimage.rotate(image_edge, 45, reshape=True)
plt.imshow(rotate_45)
```
